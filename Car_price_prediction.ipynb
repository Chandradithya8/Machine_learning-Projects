```
import pandas as pd
df=pd.read_csv("/content/carprices.csv.txt")
df
```

```
a=pd.get_dummies(df.Carmodel)
a
```

```
b=pd.concat([df,a],axis='columns')
b
```

```
c=b.drop(['Carmodel','BMW X5'],axis='columns')
c
```

```
x=c.drop(['Sell Price($)'],axis='columns')
x
```

```
y=c['Sell Price($)']
y
```

```
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```
reg.fit(x,y)
```

```
reg.predict([[45000,4,0,1]]) # Benz with 45000 mileage and 4 yrs old
```

```
reg.predict([[86000,7,0,0]]) # BMW with 86000 mileage and 7 yrs old
```

```
reg.score(x,y) # accuracy of the model
```
